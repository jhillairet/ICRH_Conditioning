```python
import ICRH_FastData as fastacq 
import ICRH_FileIO as io
import matplotlib.pylab as plt
%matplotlib notebook
```

```python
# Returns the list of all the files located in the Fast Data directory
file_list = io.list_local_files(local_data_path='data/Fast_Data')
# Then deduces the shot numbers represented by those files
fastacq.get_shot_list(file_list)
```

```python
# generate a data structure containing all the fast data of a shot
shot = 80

fd = fastacq.FastData(shot)

fd.Q1_amplitude.head()
```
